```python
import matplotlib.pyplot as plt
import numpy as np
```

# Types of graph
1. plot ---        line graph,between two parameters 
2. bar ---         bar graph of single parameters
3. hist ---        histogram of a bunch of data i.e. frequency of single param
4. scatter ---     dot graph between two param
5. boxplot/box---  statstical overview in form of box for a single param 
                   mean,outliners,qunatiles(25%,75%),IQR
                   (optional - categorical data) 
6. stem---         more like bar graph ,lines instead of bars
7. violinplot---   violin in shape statistical overview graph
                   sperad of data (more like hist)
                   single parameter (optional cat-data)

```python
plt.bar(x,y)
plt.title('bar graph')
plt.show()

plt.hist(x)
plt.title('histogram')
plt.show()

plt.plot(x,y)
plt.title('plot graph')
plt.grid()
plt.xlabel('x-->')
plt.ylabel('y-->')
plt.xticks(x)
plt.yticks(range(0,200,10))
plt.show()

plt.scatter(x,y)
plt.title('scatter')
plt.show()

plt.boxplot(x)
plt.title('boxplot graph')
plt.show()

plt.stem(x,y)
plt.title('stem graph')
plt.show()

plt.violinplot(x)
plt.title('violen plot')
plt.show()
```

```python
y2=x**4
```

```python
plt.plot(x,y,x,y2)
plt.legend(['y=x**3','y2=x**4'])
plt.show()
```

```python
plt.plot(x,y)
plt.plot(x,y2)
plt.legend(['y=x**3','y2=x**4'])
```

```python
fig=plt.figure(figsize=(10,5))
plt.plot(x,y,lw=3,marker='o',markersize='10',markerfacecolor='yellow')
plt.plot(x,y2,marker='v',markersize=10)
plt.legend(['y=x**3','y2=x**4'])
plt.grid()
plt.show()
```

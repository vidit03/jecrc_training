```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df1 = pd.read_csv('df1',index_col=0)
df2 = pd.read_csv("df2")
df3 = pd.read_csv("df3")
```

#### Graph function 
1. plot                    -simple line plot
2. plot.area               -filled version of line plot
3. plot/bar plot.barh      -bar graph
4. plot.box                -box plot
5. plot.density            -density line(smooth curved)
6. plot.hexbin             -hexagon version of scatter with density
7. plot.hist               -histogram
8. plot.kde                -kernel density estimator (density graph)
9. plot.pie                -pie chart of categorical data
10. plot.scatter           -scatter graph

```python
fig = plt.figure(figsize=[20,20])
df2.plot();plt.title('line plot');plt.show()
df2.plot.area();plt.title('area plot');plt.show()
df2.plot.bar();plt.title('bar plot');plt.show()
df2.plot.box();plt.title('box plot');plt.show()
df2.plot.barh();plt.title('bar horizontal plot');plt.show()
df2.plot.density();plt.title('density plot');plt.show()
df2.plot.kde();plt.title('kde plot');plt.show()
df2.plot.hist();plt.title('histogram plot');plt.show()
df2.plot.scatter(x='a',y='b');plt.title('scatter plot');plt.show()
df2.plot.hexbin(x='a',y='b',C='c',gridsize=6);plt.title('hexbin plot');plt.show()
```

```python
import seaborn as sns
```

#### categories
1. categorical plots(hue can be used comparison with categorical data)
     a. countplot(single param)
     b. barplot(one categorical and one numerical )
     c. boxplot
     d. violin plot
     e. swarm plot
2. distributin plots
     a. dist plot
     b. displot
     c. jointplot
     d. pair plot
     e. kdeplot
     f. rug plot
     g. scatter plot
3. matrix plots
     a. heat map
     b. cluster map
     c. pair plot (same as above one)
4. regression plots
     a. lm plot
     b. reg plot
5. grid plots
     a. pair grid
     b. joint grid
     c. facet grid

```python
sns.countplot(x='day',data=tips,hue='sex')
```

```python
sns.barplot(x='day',y='total_bill',data=tips)
```

```python
sns.barplot(x='day',y='total_bill',data=tips,estimator=sum)
```

```python
sns.barplot(x='day',y='total_bill',data=tips,hue='time',estimator=np.mean)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='time')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='time')
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips,hue='time')
```

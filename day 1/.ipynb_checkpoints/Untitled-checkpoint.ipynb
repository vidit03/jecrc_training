```python
import numpy as np
```

```python
import pandas as pd
```

```python
data = {'country':['india','usa','japan','china','india','usa'],
         'year':['2010','2010','2010','2015','2015','2015'],
         'continent':['asia','na','asia','asia','asia','na'],
         'value':[20,10,40,50,60,55]}
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
pd.pivot_table(df,index=['year'],aggfunc='sum')
```

```python
pd.pivot_table(df,index=['continent'],aggfunc='sum')
```

# concat

```python
people1 = {'id':[1,2,3,4],
           'name':['ab','rf','kl','ed'],
           'age':[45,25,41,30]}

people2 = {'id':[11,22,33,44],
           'name':['acb','rwf','kgl','edd'],
           'age':[46,23,42,31]}


people3 = {'id':[13,21,32,45],
           'name':['awb','ref','krl','etd'],
           'age':[48,27,46,35]}
```

```python
df1=pd.DataFrame(people1)
df2=pd.DataFrame(people2)
df3=pd.DataFrame(people3)
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
pd.concat([df1,df2,df3])
```

```python
pd.concat([df1,df2,df3],ignore_index=True)
```

```python
pd.concat([df1,df2,df3],axis=1)
```

```python
people4 = {'id':[1,2,3,4],
           'name':['ab','rf','kl','ed'],
           'age':[45,25,41,30]}

people5 = {'id':[1,2,3,4],
           'name':['ab','rf','kl','ed'],
           'age':[467,234,423,312]}
```

```python
df4=pd.DataFrame(people4)
df5=pd.DataFrame(people5)
```

```python
df4
```

```python
df5
```

```python
pd.merge(df4,df5,on=['id'])
```

```python
people6 = {'id':[1,2,3,4],
           'name':['ab','rf','kl','ed'],
           'age':[45,25,41,30]}

people7 = {
           's_name':['b','f','l','d'],
           'ph':[4677876,6587587234,4655723,376876687612]}
```

```python
df6=pd.DataFrame(people6)
df7=pd.DataFrame(people7)
```

```python
df6.join(df7)
```
